```python
with open('fra.txt', 'rb') as f:
    string = f.read().decode()
```

```python
lines = string.split('\n')
```

```python
linelist = []
linelist = list(map(lambda x: tuple(x.split("\t")), lines))
```

```python
linelist[50:100]
```

```python
import pandas as pd
```

```python
columns = ["English", "French"]
df = pd.DataFrame(linelist, columns=['English', 'French'])
```

```python
df.tail()
```

```python
df.to_csv("english_to_french.csv", index=False)
```
